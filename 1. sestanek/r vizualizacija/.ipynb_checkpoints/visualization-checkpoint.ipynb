```r
library(igraph)
library(dplyr)
```

```r
nodes <- read.csv("vozli.csv", header=T, as.is=T)
links <- read.csv("povezave.csv", header=T, as.is=T)
kratice <- read.csv("kratice.csv", header=T, as.is=T)
```

```r
links$norm <- scale(links$volume) + 2
```

```r
net <- graph_from_data_frame(d=links, vertices=nodes, directed=T) 
net
```

```r
nodes2 <- read.csv("vozli2.csv", header=T, as.is=T)
links2 <- read.csv("povezave2.csv", header=T, as.is=T)
links2 <- inner_join(links2, kratice, by='borza')
links2$norm <- scale(links2$volume) + 2
net2 <- graph_from_data_frame(d=links2, vertices=nodes2, directed=F)
colrs <- c("tomato", "gold")
V(net2)$color <- colrs[V(net2)$currencyType]


plot(net2, edge.width=links2$norm, edge.color = links2$barva, layout=layout_nicely)
legend('topleft', unique(links2$borza), pch=21,
       col="#777777", pt.bg=unique(links2$barva), pt.cex=2, cex=.8, bty="n", ncol=1)
```

```r
vizualizacija <- function(vozli_csv, povezave_csv, kratice_csv) {
    nodes2 <- read.csv(vozli_csv, header=T, as.is=T)
    links2 <- read.csv(povezave_csv, header=T, as.is=T)
    kratice <- read.csv(kratice_csv, header=T, as.is=T)
    links2 <- inner_join(links2, kratice, by='borza')
    links2$norm <- scale(links2$volume) + 2
    net2 <- graph_from_data_frame(d=links2, vertices=nodes2, directed=F)
    colrs <- c("tomato", "gold")
    V(net2)$color <- colrs[V(net2)$currencyType]
    plot(net2, edge.width=links2$norm, edge.color = links2$barva, layout=layout_nicely)
    legend('topleft', unique(links2$borza), pch=21,
       col="#777777", pt.bg=unique(links2$barva), pt.cex=2, cex=.8, bty="n", ncol=1)
}
```

```r
vizualizacija('vozli.csv', 'povezave.csv', 'kratice.csv')
```
