<b>Končni rezultat:</b>  Podobno se lahko stori za vse preostale večje borze. Na koncu se dobljene podatke lahko združi v celoto.

<b>Primer zajetih podatkov:</b>

Pomen stolpcev:
    - date (unix format)
    - tid: id transakcije
    - price: cena izražena v enotah druge valute v paru
    - amount: količina tokenov v transakciji
    - type: tip transakcije, 0 pomeni 'buy', 1 pomeni 'sell'

Te informacije sem zajel o vseh trgovalnih parih na Bitstampu (ne le za par 'btceur').

```python
bitstamp_raw = pd.read_csv('btceur.csv')
bitstamp_raw.head()
```

<b>Primer obdelanih podatkov:</b>

Pomen stolpcev:
    - volume: dnevni promet na izbrani relaciji 
    - flow: gre v obe smeri, tako je npr. ločen 'BTCETH' in 'ETHBTC'
    - nr_tokens: število kovancev/tokenov s katerimi se je trgovalo na dani relaciji
    - mean_daily_price_$: povprečena dnevna cena kriptovalute v paru (če sta v paru dve kriptovaluti, gre za ceno druge)
    - nr_txs: število transakcij 


```python
bitstamp = pd.read_csv('tidy.csv')
bitstamp.columns = ['volume', 'flow', 'nr_tokens', 'mean_daily_price_$', 'nr_txs', 'prva', 'druga']
bitstamp
```
